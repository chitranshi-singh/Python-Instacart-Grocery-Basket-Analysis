## 04. Transposing

```python
df_dep.head()
```

```python
# transposing df_dep
df_dep.T
```

```python
# creating new dataframe from the transposed version
df_dep_t=df_dep.T
```

```python
df_dep_t
```

```python
# adding index to df_dep_t
df_dep_t.reset_index()
```

```python
# take the first row of df_dep_t for the header
new_header=df_dep_t.iloc[0]
```

```python
new_header
```

```python
# take the data under the header row for a new df
df_dep_t_new=df_dep_t[1:]
```

```python
df_dep_t_new
```

```python
# setting the header row as the df header
df_dep_t_new.columns=new_header
```

```python
df_dep_t_new
```
