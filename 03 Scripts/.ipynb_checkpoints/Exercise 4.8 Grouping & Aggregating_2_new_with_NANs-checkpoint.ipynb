# Create a spending flag for each user based on the avg price of all their orders

```python
# create avg_spending column by splitting data into user_id groups and applying transform function on prices
df_ords_prods_merged['avg_spending']=df_ords_prods_merged.groupby('user_id')['prices'].transform(np.mean)
```

```python
# check the output
df_ords_prods_merged.head()
```

```python
# create a spending_flag for 'Low spender' using loc
df_ords_prods_merged.loc[df_ords_prods_merged['avg_spending']<10, 'spending_flag']='Low spender'
```

```python
# create a spending_flag for 'High spender' using loc
df_ords_prods_merged.loc[df_ords_prods_merged['avg_spending']>=10, 'spending_flag']='High spender'
```

```python
# check the frequency of spending_flag
df_ords_prods_merged['spending_flag'].value_counts(dropna=False)
```

```python
pd.options.display.max_rows = None
```

```python
# check the first 100 rows of user_id, spending_flag and avg_order_price columns
df_ords_prods_merged[['user_id', 'spending_flag', 'avg_spending']].head(100)
```

# Create an order frequency flag for each user

```python
# create median column by splitting data into user_id groups and applying transform function on days_since_prior_order
df_ords_prods_merged['order_frequency']=df_ords_prods_merged.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
# check the output
df_ords_prods_merged.head()
```

```python
# create an order_frequency_flag for 'Non-frequent customer' flag using loc
df_ords_prods_merged.loc[df_ords_prods_merged['order_frequency']>20, 'order_frequency_flag']='Non-frequent customer'
```

```python
# create an order_frequency_flag for 'Regular customer' flag using loc
df_ords_prods_merged.loc[(df_ords_prods_merged['order_frequency']<=20) & (df_ords_prods_merged['order_frequency']>10), 'order_frequency_flag']='Regular customer'
```

```python
# create an order_frequency_flag for 'Frequent customer' flag using loc
df_ords_prods_merged.loc[df_ords_prods_merged['order_frequency']<=10, 'order_frequency_flag']='Frequent customer'
```

```python
# check the frequency of order_frequency_flag
df_ords_prods_merged['order_frequency_flag'].value_counts(dropna=False)
```

```python
# check the first 100 rows of user_id, spending_flag and avg_order_price columns
df_ords_prods_merged[['user_id', 'order_frequency_flag', 'order_frequency']].head(100)
```
