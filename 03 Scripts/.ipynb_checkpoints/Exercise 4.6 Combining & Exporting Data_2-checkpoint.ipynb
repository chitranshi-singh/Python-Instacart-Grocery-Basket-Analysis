# Merging Data

```python
# drop the previously existing merge flag from df_ords_prods_combined
df_ords_prods_combined=df_ords_prods_combined.drop(columns='_merge')
```

```python
# check the output
df_ords_prods_combined.head()
```

```python
# merge df_prods and df_ords_prods_combined using product_id as key, add a merge flag
df_merged=df_prods.merge(df_ords_prods_combined, on='product_id', indicator=True)
```

```python
# check the output
df_merged.head()
```

```python
# check the frequency of the merge flag
df_merged['_merge'].value_counts()
```

```python
# check the total size of df_merged
df_merged.shape
```

# Exporting Data

```python
# export data to pkl
df_merged.to_pickle(os.path.join(path,'02 Data','Prepared Data','orders_products_merged.pkl'))
```
