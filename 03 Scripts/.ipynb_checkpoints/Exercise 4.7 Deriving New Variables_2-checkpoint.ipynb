## Update busiest_day to busiest_days

```python
# rename busiest_day to busiest_days
df_ords_prods_merged.rename(columns={'busiest_day':'busiest_days'}, inplace=True)
```

```python
# check the output
df_ords_prods_merged.head()
```

```python
# create a variable to store frequency of orders_day_of_week
orders_day_frequency=df_ords_prods_merged['orders_day_of_week'].value_counts(dropna=False)
```

```python
# check the output
orders_day_frequency
```

```python
# convert index values of orders_day_frequency into list
orders_day_list=list(orders_day_frequency.index)
```

```python
# check the output
orders_day_list
```

```python
# run a for-loop
result=[]

for value in df_ords_prods_merged['orders_day_of_week']:
    if (value==orders_day_list[0]) | (value==orders_day_list[1]):
        result.append('Most busy')
    elif (value==orders_day_list[-1]) | (value==orders_day_list[-2]):
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# check the output
result
```

```python
# check the length of result
len(result)
```

```python
# check the total size of df_ords_prods_merged
df_ords_prods_merged.shape
```

```python
# add values of result to busiest_days
df_ords_prods_merged['busiest_days']=result
```

```python
# check the frequency of busiest_days
df_ords_prods_merged['busiest_days'].value_counts(dropna=False)
```

```python
# check df_ords_prods_merged
df_ords_prods_merged.head()
```

```python
# check the total size of df_ords_prods_merged
df_ords_prods_merged.shape
```

## Create column busiest_period_of_day

```python
# create a variable to store frequency of order_hour_of_day as dictionary
order_hour_dict=df_ords_prods_merged['order_hour_of_day'].value_counts(dropna=False).to_dict()
```

```python
# check the output
order_hour_dict
```

```python
# create a variable to calculate quantile of order_hour_of_day frequency
order_hour_quantile=df_ords_prods_merged['order_hour_of_day'].value_counts(dropna=False).quantile([.75, .25])
```

```python
# check the output
order_hour_quantile
```

```python
# convert order_hour_quantile into list
order_hour_list=list(order_hour_quantile)
```

```python
# check the output
order_hour_list
```

```python
# run a for-loop
result_1=[]

for value in df_ords_prods_merged['order_hour_of_day']:
    for key, val in order_hour_dict.items():
        if value==key:
            if val>order_hour_list[0]:
                result_1.append('Most orders')
            elif val<=order_hour_list[1]:
                result_1.append('Fewest orders')
            else:
                result_1.append('Average orders')
```

```python
# check the output
result_1
```

```python
# check the length of result_1
len(result_1)
```

```python
# check the total size of df_ords_prods_merged
df_ords_prods_merged.shape
```

```python
# add values of result_1 to a new column
df_ords_prods_merged['busiest_period_of_day']=result_1
```

```python
# check the frequency of busiest_period_of_day
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna=False)
```

```python
# check df_ords_prods_merged
df_ords_prods_merged.head()
```
