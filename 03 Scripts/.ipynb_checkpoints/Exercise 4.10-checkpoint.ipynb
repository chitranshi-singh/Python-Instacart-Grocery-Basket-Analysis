## Create customer profiles

```python
# check frequency of family_status
df_final['family_status'].value_counts(dropna=False)
```

```python
# check frequency of dependants
df_final['n_dependants'].value_counts(dropna=False)
```

```python
# check basic stats of dependants for each family_status
df_final.groupby('family_status').agg({'n_dependants':['mean', 'min', 'max']})
```

```python
# create a list variable to calculate quantile of income
income_list=list(df_final['income'].quantile([.15, .90]))
```

```python
# check the output
income_list
```

```python
# create customer profile for retired individuals

df_final.loc[(df_final['age']>65) & (df_final['income']<=income_list[0]), 'customer_profile']='retired low-income'
df_final.loc[(df_final['age']>65) & (df_final['income']>income_list[0]) & (df_final['income']<=income_list[1]), 'customer_profile']='retired avg-income'
df_final.loc[(df_final['age']>65) & (df_final['income']>income_list[1]), 'customer_profile']='retired high-income'
```

```python
# create customer profile for single individuals

df_final.loc[(df_final['age']<=65) & (df_final['n_dependants']==0) & (df_final['income']<=income_list[0]), 'customer_profile']='single low-income'
df_final.loc[(df_final['age']<=65) & (df_final['n_dependants']==0) & (df_final['income']>income_list[0]) & (df_final['income']<=income_list[1]), 'customer_profile']='single avg-income'
df_final.loc[(df_final['age']<=65) & (df_final['n_dependants']==0) & (df_final['income']>income_list[1]), 'customer_profile']='single high-income'
```

```python
# create customer profile for families

df_final.loc[(df_final['age']<=65) & (df_final['n_dependants']>0) & (df_final['income']<=income_list[0]), 'customer_profile']='family low-income'
df_final.loc[(df_final['age']<=65) & (df_final['n_dependants']>0) & (df_final['income']>income_list[0]) & (df_final['income']<=income_list[1]), 'customer_profile']='family avg-income'
df_final.loc[(df_final['age']<=65) & (df_final['n_dependants']>0) & (df_final['income']>income_list[1]), 'customer_profile']='family high-income'
```

```python
# check the output
df_final['customer_profile'].value_counts(dropna=False)
```

### Are there differences in ordering habits based on a customerâ€™s loyalty status?

```python
# create and export a pie chart of loyalty status

myexplode = [0.1, 0.03, 0.03]
mycolors=['darkorange', 'orange', 'navajowhite']
mytextprops = {'size':9}
mylabels=['Regular customer (10-40 orders)', 'Loyal customer (>40 orders)', 'New customer (<10 orders)']
font1 = {'family':'serif','color':'black','size':15}

loyalty_status_pie = df_final['loyalty_flag'].value_counts().plot(kind='pie', figsize=(10,6), counterclock=False, startangle=-90, y='user_id', ylabel='', autopct='%1.0f%%', legend='', explode=myexplode, labels=mylabels, colors=mycolors)
plt.title('Distribution of loyalty status', fontdict=font1)

loyalty_status_pie.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Final viz', 'pie_loyalty_status.png'))
```

```python
# create a crosstab of order frequency by loyalty status
order_freq_by_loyalty=pd.crosstab(df_final['loyalty_flag'], df_final['order_frequency_flag']).apply(lambda r: r/r.sum()*100, axis=1)
```

```python
# check the output
order_freq_by_loyalty
```

```python
# create and export a 100% stacked bar chart of order frequency by loyalty status

mycolor=['darkgreen', 'forestgreen', 'mediumseagreen']

order_freq_by_loyalty_bar=order_freq_by_loyalty.plot(kind='barh', stacked=True, color=mycolor, figsize=(10,5.5))

font1 = {'family':'serif','color':'black','size':14}
plt.title('Order frequency by loyalty status', fontdict=font1)
plt.ylabel('')
order_freq_by_loyalty_bar.legend(['Frequent customers (<10 days)', 'Non-frequent customers (<20 days)', 'Regular customer (10-20 days)'], bbox_to_anchor=(1.0, 1.0), edgecolor = 'darkorange')

df_total = order_freq_by_loyalty["Frequent customer"] + order_freq_by_loyalty["Non-frequent customer"] + order_freq_by_loyalty["Regular customer"]
df_rel = order_freq_by_loyalty[order_freq_by_loyalty.columns[0:]].div(df_total, 0) * 100

for n in df_rel:
    for i, (cs, ab, pc) in enumerate(zip(order_freq_by_loyalty.iloc[:, 0:].cumsum(1)[n],
                                         order_freq_by_loyalty[n], df_rel[n])):
        plt.text(cs - ab / 2, i, str(round(pc)) + '%',
                 va='center', ha='center', rotation=0, fontsize=8)
        
plt.tight_layout()

order_freq_by_loyalty_bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Final viz', 'bar_order_freq_by_loyalty.png'))
```

```python
# create a new variable to store departments for each customer profile
cust_dept=df_final.groupby('customer_profile')['department'].value_counts()
```

```python
# check the output
cust_dept
```

```python
# create a dict consisting of customer_profile as key and dict of dept name (key) and dept frequency (value) as value

cust_dept_dict = {}
for k, v in cust_dept.items():
    if cust_dept_dict == {} or cust_dept_dict.get(k[0]) is None:
        temp_dict = {k[1]: v}
        cust_dept_dict[k[0]] = temp_dict
    else:
        current_dict = (cust_dept_dict.get(k[0]))
        key = k[1]
        value = v
        current_dict[key] = value
```

```python
# sort cust_dept_dict by dept frequency

sorted_dict_by_dept_score = {}
for x, y in cust_dept_dict.items():
    sorted_dept = sorted(y.items(), key=lambda x: x[1], reverse=True)
    sorted_dict_by_dept_score[x] = sorted_dept
```

```python
# calculate top 5 dept for each customer_profile from cust_dept_dict

top_5_dept_dict = {}
for x, y in sorted_dict_by_dept_score.items():
    top_5_dept_dict[x] = y[0:5]
```

```python
# check the output
top_5_dept_dict
```

```python
# calculate percentage of dept frequency for each customer_profile

percentile_dict = {}
for key, val in top_5_dept_dict.items():
    sum = 0;
    new_dict_val = []
    for pair_val in val:
        sum = sum + pair_val[1]
    for pair_val in val:
        percentile = pair_val[1] / sum * 100
        temp_list = list(pair_val)
        temp_list.append(percentile)
        temp_toup = tuple(temp_list)
        new_dict_val.append(temp_toup)
        percentile_dict[key] = new_dict_val
```

```python
# check the output
percentile_dict
```

```python
# convert percentile_dict to a list of lists where each list element contains customer_profile, dept name, dept frequency percentage

final_list = []
for outerKey, outerVal in percentile_dict.items():
    for innerValue in outerVal:
        tempList = []
        tempList.append(outerKey)
        tempList.append(innerValue[0])
        tempList.append(innerValue[1])
        tempList.append(innerValue[2])
        final_list.append(tempList)
```

```python
# check the output
final_list
```

```python
# convert list to dataframe
df_new = pd.DataFrame(final_list, columns=['customer_profile', 'department', 'department_score', 'department_percentile'])
```

```python
# check the output
df_new
```

```python
# create and export a 100% stacked bar chart of top 5 dept by customer profile

mycolor=['darkgreen', 'forestgreen', 'mediumseagreen', 'darkseagreen', 'lightgreen']
font1 = {'family':'serif','color':'black','size':14}

cust_top_dept_bar=df_new.groupby(['customer_profile', 'department'], sort=False)['department_percentile'].sum().unstack().plot(kind='barh', stacked=True, color=mycolor, figsize=(8,5))

plt.title('Top 5 departments by customer profile', fontdict=font1)
cust_top_dept_bar.legend(bbox_to_anchor=(1.0, 1.0), edgecolor = 'darkorange')
plt.xlabel('(in %)')
plt.ylabel('')

plt.tight_layout()

cust_top_dept_bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Final viz', 'bar_cust_profile_top_dept.png'))
```
