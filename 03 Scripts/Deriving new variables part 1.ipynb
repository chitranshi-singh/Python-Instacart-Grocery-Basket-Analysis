# 02. Create price_label column

```python
# create condition for 'high-range product'
df_ords_prods_merged.loc[df_ords_prods_merged['prices']>15, 'price_label']='High-range product'
```

```python
# create condition for 'mid-range product'
df_ords_prods_merged.loc[(df_ords_prods_merged['prices']<=15) & (df_ords_prods_merged['prices']>5), 'price_label']='Mid-range product'
```

```python
# create condition for 'low-range product'
df_ords_prods_merged.loc[df_ords_prods_merged['prices']<=5, 'price_label']='Low-range product'
```

```python
# check the frequency of new column price_label
df_ords_prods_merged['price_label'].value_counts(dropna=False)
```

# 03. Create busiest_day column

```python
# check the frequency of orders_day_of_week
df_ords_prods_merged['orders_day_of_week'].value_counts()
```

```python
# run a for-loop
result=[]

for value in df_ords_prods_merged['orders_day_of_week']:
    if value==0:
        result.append('Busiest day')
    elif value==4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# check the values of result column
result
```

```python
# create a new column equal to values of result column
df_ords_prods_merged['busiest_day']=result
```

```python
# check the frequency of busiest_day
df_ords_prods_merged['busiest_day'].value_counts(dropna=False)
```
