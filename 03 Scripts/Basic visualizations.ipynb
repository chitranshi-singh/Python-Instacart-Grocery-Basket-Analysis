## 03. Visualizations

## a) Bar chart

```python
# create a bar chart: orders_day_of_week
df_ords_prods_cust['orders_day_of_week'].value_counts().plot.bar()
```

```python
# sort the bar chart to its index order
bar_1=df_ords_prods_cust['orders_day_of_week'].value_counts().sort_index().plot.bar(rot=0)
plt.xlabel('Day of week')
plt.ylabel('Total orders (Million)')
plt.title('Total orders placed per day of week')
```

```python
# export bar_1 as png
bar_1.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_orders_day_of_week.png'))
```

```python
# create a bar chart: loyalty_flag
bar_2=df_ords_prods_cust['loyalty_flag'].value_counts().plot.bar(rot=0)
plt.ylabel('Total customers (Million)')
plt.title('Customer classification')
```

```python
# export bar_2 as png
bar_2.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_loyalty_flag.png'))
```

## b) Histogram

```python
# create a histogram: prices
hist_1=df_ords_prods_cust['prices'].plot.hist(bins=70)
plt.xlabel('Prices (USD)')
plt.ylabel('Total products (Million)')
plt.title('Total products by prices')
```

```python
# export hist_1 as png
hist_1.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'hist_prices_frequency.png'))
```

```python
# create a histogram: order_hour_of_day
hist_2=df_ords_prods_cust['order_hour_of_day'].plot.hist(bins=24)
plt.xlabel('Hour of day')
plt.ylabel('Total orders (Million)')
plt.title('Total orders placed per hour of day')
```

#### The above histogram demonstrates that sales gradually start peaking from 6, with 10 to 15 being the busiest hours of the day, after which they begin to progressively decline. The hours between 0 to 5 have the least sales.

```python
# export hist_2 as png
hist_2.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'hist_order_hour_of_day_freq.png'))
```

## c) Scatterplot

```python
# create a scatterplot: prices
splot_1=sns.scatterplot(x='prices', y='prices', data=df_ords_prods_cust)
plt.xlabel('')
plt.ylabel('')
plt.suptitle('Distribution of prices')
plt.title('(USD)', fontsize=10)
```

```python
# export splot_1 as png
splot_1.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'splot_prices.png'))
```

```python
# create a scatterplot: age vs. income
splot_2=sns.scatterplot(x='age', y='income', data=df_ords_prods_cust)
plt.xlabel('Age')
plt.ylabel('Total income (USD)')
plt.title('Relationship between age and income')
```

#### The above scatterplot demonstrates a positive relationship between age and income. As age increases, so does income. The spending power of individuals above 40 years is significantly higher as opposed to those less than 40 years, however this group constitutes a rather small proportion. The mass of individuals aged 40 years and less have an average income of around 100-200k, while those more than 40 have an average income of around 100-300k, making up a majority of the proportion.

```python
# export splot_2 as png
splot_2.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'splot_age_income.png'))
```

## d) Line chart

### prices vs. orders_day_of_week

```python
# set the seed
np.random.seed(4)
```

```python
# create a list holding True/False values to the test np.random.rand() <= 0.7
dev=np.random.rand(len(df_ords_prods_cust))<=0.7
```

```python
# check the output
dev
```

```python
# store 70% of the sample in the dataframe big
big=df_ords_prods_cust[dev]
```

```python
# store 30% of the sample in the dataframe small
small=df_ords_prods_cust[~dev]
```

```python
# check the row count of df_ords_prods_cust
len(df_ords_prods_cust)
```

```python
# check the row count of the big and small dataframes
len(big) + len(small)
```

```python
# create a new subset from the small subset
df_2=small[['orders_day_of_week', 'prices']]
```

```python
# check the output
df_2.head()
```

```python
# create a line chart
line_1=sns.lineplot(data=df_2, x='orders_day_of_week', y='prices')
plt.xlabel('Day of week')
plt.ylabel('Prices (USD)')
plt.title('Expenditure per day of week')
```

```python
# export line_1 as png
line_1.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'line_prices_day_of_week.png'))
```
